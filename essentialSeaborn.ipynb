```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt  # Add this for axes control

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Scatter plot
sns.scatterplot(x=x, y=y, color='blue', marker='o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Line plot
sns.lineplot(x=x, y=y, color='green')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# Bar plot
categories = ['Category A', 'Category B', 'Category C']
values = [30, 45, 20]
sns.barplot(x=categories, y=values, color='purple')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt  # This is necessary for titles, labels, etc.

# Create a histogram
data = np.random.randn(1000)
sns.histplot(data, bins=30, color='orange', kde=True)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Create a kernel density estimation (KDE) plot
sns.kdeplot(data, color='red')
plt.title('KDE Plot')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()

# Create a rug plot
sns.rugplot(data, height=0.2, color='green')
plt.title('Rug Plot')
plt.xlabel('Values')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt  # This is necessary for titles, labels, etc.

# Create a box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
sns.boxplot(data=data, palette='pastel')
plt.title('Box Plot')
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.show()

# Create a violin plot
sns.violinplot(data=data, inner='quartile', palette='pastel')
plt.title('Violin Plot')
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.show()

# Create a swarm plot
sns.swarmplot(data=data, color='purple', size=3)
plt.title('Swarm Plot')
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.show()
```
